```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('retail_sales_dataset.csv')
df.head()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

### convert type of Column Date

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)
```

```python
df['Date'].unique()
```

```python
plt.figure(figsize=(5,5))
qual_palette = sns.color_palette("pastel",4)
sns.barplot(x=df['Product Category'],y=df['Price per Unit'],palette=qual_palette)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
qual_palette = sns.color_palette("pastel",4)
sns.barplot(x=df['Product Category'],y=df['Quantity'],palette=qual_palette)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(5,5))
qual_palette = sns.color_palette("pastel",4)
sns.barplot(x=df['Gender'],y=df['Total Amount'],palette=qual_palette)
plt.grid()
plt.show()
```

```python
gander=df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].astype('category').cat.codes
```

```python
hex = sns.color_palette(["#FF6B6B", "#4ECDC4", "#45B7D1","#FAD02E","#594F4F"])
sns.countplot(x=gander.index,palette=hex,)
plt.show()
```

```python
top_10_date=df['Date'].value_counts().head(10)
```

```python
top_10_date.index
```

```python
top_10_total_amounts=df['Total Amount'].value_counts().head(10)
```

```python
top_10_total_amounts.index
```

```python
plt.figure(figsize=(5,5))
plt.title('relationship top_10_date and top_10_total_amounts')
qual_palette = sns.color_palette("pastel",4)
sns.barplot(x=top_10_date.index,y=top_10_total_amounts.index,palette=qual_palette)
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
